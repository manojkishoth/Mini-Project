```python
r=6371

def DistanceHaversine(bus, student):
    # Convert degrees to radians
    coordinates = bus.LATITUDE, bus.LONGITUDE, student.LATITUDE, student.LONGITUDE
    # radians(c) is same as c*pi/180
    phi1, lambda1, phi2, lambda2 = [
        radians(c) for c in coordinates
    ]

    # Apply the haversine formula
    a = (np.square(sin((phi2-phi1)/2)) + cos(phi1) * cos(phi2) *
         np.square(sin((lambda2-lambda1)/2)))
    d = 2*r*asin(np.sqrt(a))
    return d
```

```python
from geopy.distance import geodesic as GD

def DistanceGeodesic(bus, student):
    a = (bus.LATITUDE, bus.LONGITUDE)
    b = (student.LATITUDE, student.LONGITUDE)
    d = GD(a,b).km
    return d
```

```python
from geopy.distance import great_circle as GRC

def DistanceGreatCircle(bus, student):
    a = (bus.LATITUDE, bus.LONGITUDE)
    b = (student.LATITUDE, student.LONGITUDE)
    d = GRC(a,b).km
    return d
```
